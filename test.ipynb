```python
import matplotlib.pyplot as plt

# Data for linear graph
x1 = [1, 2, 3, 4, 5]
y1 = [2, 8, 27, 64, 125] # cubic relation y=x^3
x=[1,2,6,7,8]  
y2 = [(2*i +1) for i in x]
# Plot the graph
plt.plot(x1, y1, label="y = x^3", color="blue", marker="o")
plt.plot(x,y2,label="y=2x+1",color="red",marker="s")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Linear Graph using Matplotlib")
plt.legend()

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# X values (common for all functions)
x = [0, 1, 2, 3, 4, 5]

# Define multiple linear functions
y1 = [2*i for i in x]       # y = 2x
y2 = [3*i + 1 for i in x]   # y = 3x + 1
y3 = [i - 2 for i in x]     # y = x - 2

# Plot all lines
plt.plot(x, y1, label="y = 2x", color="blue", marker="o")
plt.plot(x, y2, label="y = 3x + 1", color="green", marker="s")
plt.plot(x, y3, label="y = x - 2", color="red", marker="^")

# Add labels, title, grid
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Linear Functions")
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
```
